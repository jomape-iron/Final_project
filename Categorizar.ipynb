```python
import pandas as pd
import numpy as np
```

```python
# Cargar los DataFrames desde archivos pickle
df_absentismo = pd.read_pickle('df_absentismo.pkl')
df_historical_performance = pd.read_pickle('df_historical_performance.pkl')
df_performance = pd.read_pickle('df_performance.pkl')
df_actual = pd.read_pickle('df_actual.pkl')
df_base = pd.read_pickle('df_base.pkl')
```

```python
# Crear un diccionario para la categorización
categorization = {}

# Categorizar la columna 'IT/AT'
it_at_categories = df_absentismo['IT/AT'].unique()
it_at_mapping = {category: i + 1 for i, category in enumerate(it_at_categories)}
categorization['IT/AT'] = it_at_mapping
df_absentismo['IT/AT'] = df_absentismo['IT/AT'].map(it_at_mapping)

# Categorizar la columna 'Horas contratadas'
horas_contratadas_categories = df_absentismo['Horas contratadas'].unique()
horas_contratadas_mapping = {category: i + 1 for i, category in enumerate(horas_contratadas_categories)}
categorization['Horas contratadas'] = horas_contratadas_mapping
df_absentismo['Horas contratadas'] = df_absentismo['Horas contratadas'].map(horas_contratadas_mapping)

# Categorizar la columna 'TIPO DE BAJA'
tipo_baja_categories = df_absentismo['TIPO DE BAJA'].unique()
tipo_baja_mapping = {category: i + 1 for i, category in enumerate(tipo_baja_categories)}
categorization['TIPO DE BAJA'] = tipo_baja_mapping
df_absentismo['TIPO DE BAJA'] = df_absentismo['TIPO DE BAJA'].map(tipo_baja_mapping)

# Crear un nuevo DataFrame con las columnas categorizadas y la columna 'ID'
df_absentismo_c = df_absentismo[['ID'] + list(categorization.keys())]

print(categorization)
print(df_absentismo_c)
```

```python
# Crear un diccionario para la categorización
categorization = {}

# Categorizar la columna 'Supervisor/a'
supervisor_categories = df_historical_performance['Supervisor/a'].unique()
supervisor_mapping = {category: i + 1 for i, category in enumerate(supervisor_categories)}
categorization['Supervisor/a'] = supervisor_mapping
df_historical_performance['Supervisor/a'] = df_historical_performance['Supervisor/a'].map(supervisor_mapping)

# Categorizar la columna 'Evaluación'
evaluacion_categories = df_historical_performance['Evaluación'].unique()
evaluacion_mapping = {category: i + 1 for i, category in enumerate(evaluacion_categories)}
categorization['Evaluación'] = evaluacion_mapping
df_historical_performance['Evaluación'] = df_historical_performance['Evaluación'].map(evaluacion_mapping)

# Crear un nuevo DataFrame con las variables categorizadas y la columna 'ID'
df_historical_performance_c = pd.DataFrame(df_historical_performance[['ID'] + list(categorization.keys())])

print(categorization)
print(df_historical_performance_c)
```

```python
# Crear un diccionario para la categorización
categorization = {}

# Categorizar la columna 'Evaluación'
evaluacion_categories = df_performance['Evaluación'].unique()
evaluacion_mapping = {category: i + 1 for i, category in enumerate(evaluacion_categories)}
categorization['Evaluación'] = evaluacion_mapping
df_performance['Evaluación'] = df_performance['Evaluación'].map(evaluacion_mapping)

# Categorizar la columna 'Supervisor/a'
supervisor_categories = df_performance['Supervisor/a'].unique()
supervisor_mapping = {category: i + 1 for i, category in enumerate(supervisor_categories)}
categorization['Supervisor/a'] = supervisor_mapping
df_performance['Supervisor/a'] = df_performance['Supervisor/a'].map(supervisor_mapping)

# Crear un nuevo DataFrame con las variables categorizadas y la columna 'ID'
df_performance_c = pd.DataFrame(df_performance[['ID'] + list(categorization.keys())])

print(categorization)
print(df_performance_c)
```

```python
# Create a dictionary for categorization
categorization = {}

# Categorize 'HORAS DE CONTRATO' column
horas_contrato_categories = df_actual['HORAS DE CONTRATO'].unique()
horas_contrato_mapping = {category: i + 1 for i, category in enumerate(horas_contrato_categories)}
categorization['HORAS DE CONTRATO'] = horas_contrato_mapping
df_actual['HORAS DE CONTRATO'] = df_actual['HORAS DE CONTRATO'].map(horas_contrato_mapping)

# Categorize 'Categoría' column
categoria_categories = df_actual['Categoría'].unique()
categoria_mapping = {category: i + 1 for i, category in enumerate(categoria_categories)}
categorization['Categoría'] = categoria_mapping
df_actual['Categoría'] = df_actual['Categoría'].map(categoria_mapping)

# Categorize 'Número de hijos' column
num_hijos_categories = df_actual['Número de hijos'].unique()
num_hijos_mapping = {category: i + 1 for i, category in enumerate(num_hijos_categories)}
categorization['Número de hijos'] = num_hijos_mapping
df_actual['Número de hijos'] = df_actual['Número de hijos'].map(num_hijos_mapping)

# Categorize 'Sexo' column
sexo_categories = df_actual['Sexo'].unique()
sexo_mapping = {category: i + 1 for i, category in enumerate(sexo_categories)}
categorization['Sexo'] = sexo_mapping
df_actual['Sexo'] = df_actual['Sexo'].map(sexo_mapping)

# Categorize 'Supervisora/Coordinador' column
supervisor_categories = df_actual['Supervisora/Coordinador'].unique()
supervisor_mapping = {category: i + 1 for i, category in enumerate(supervisor_categories)}
categorization['Supervisora/Coordinador'] = supervisor_mapping
df_actual['Supervisora/Coordinador'] = df_actual['Supervisora/Coordinador'].map(supervisor_mapping)

# Categorize 'Sección' column
seccion_categories = df_actual['Sección'].unique()
seccion_mapping = {category: i + 1 for i, category in enumerate(seccion_categories)}
categorization['Sección'] = seccion_mapping
df_actual['Sección'] = df_actual['Sección'].map(seccion_mapping)

# Add the 'ID' column to the categorized DataFrame
df_actual_c = pd.concat([df_actual['ID'], df_actual[list(categorization.keys())]], axis=1)

print(categorization)
print(df_actual_c)
```

```python
# Create a dictionary for categorization
categorization = {}

# Categorize 'CATEGORIA' column
categoria_categories = df_base['CATEGORIA'].unique()
categoria_mapping = {category: i + 1 for i, category in enumerate(categoria_categories)}
categorization['CATEGORIA'] = categoria_mapping
df_base['CATEGORIA'] = df_base['CATEGORIA'].map(categoria_mapping)

# Categorize 'SECCIÓN' column
seccion_categories = df_base['SECCIÓN'].unique()
seccion_mapping = {category: i + 1 for i, category in enumerate(seccion_categories)}
categorization['SECCIÓN'] = seccion_mapping
df_base['SECCIÓN'] = df_base['SECCIÓN'].map(seccion_mapping)

# Categorize 'TIENDA/SECCION INICIAL' column
tienda_seccion_inicial_categories = df_base['TIENDA/SECCION INICIAL'].unique()
tienda_seccion_inicial_mapping = {category: i + 1 for i, category in enumerate(tienda_seccion_inicial_categories)}
categorization['TIENDA/SECCION INICIAL'] = tienda_seccion_inicial_mapping
df_base['TIENDA/SECCION INICIAL'] = df_base['TIENDA/SECCION INICIAL'].map(tienda_seccion_inicial_mapping)

# Categorize 'SECCIÓN2' column
seccion2_categories = df_base['SECCIÓN2'].unique()
seccion2_mapping = {category: i + 1 for i, category in enumerate(seccion2_categories)}
categorization['SECCIÓN2'] = seccion2_mapping
df_base['SECCIÓN2'] = df_base['SECCIÓN2'].map(seccion2_mapping)

# Categorize 'TIENDA/SECCION FINAL2' column
tienda_seccion_final2_categories = df_base['TIENDA/SECCION FINAL2'].unique()
tienda_seccion_final2_mapping = {category: i + 1 for i, category in enumerate(tienda_seccion_final2_categories)}
categorization['TIENDA/SECCION FINAL2'] = tienda_seccion_final2_mapping
df_base['TIENDA/SECCION FINAL2'] = df_base['TIENDA/SECCION FINAL2'].map(tienda_seccion_final2_mapping)

# Categorize 'QUIEN CAUSO BAJA?' column
causa_baja_categories = df_base['QUIEN CAUSO BAJA?'].unique()
causa_baja_mapping = {category: i + 1 for i, category in enumerate(causa_baja_categories)}
categorization['QUIEN CAUSO BAJA?'] = causa_baja_mapping
df_base['QUIEN CAUSO BAJA?'] = df_base['QUIEN CAUSO BAJA?'].map(causa_baja_mapping)

# Add the 'ID' column to the categorized DataFrame
df_base_c = pd.concat([df_base['ID'], df_base[list(categorization.keys())]], axis=1)

# Category for "FECHA BAJA"
df_base['FECHA BAJA CATEGORY'] = np.where(df_base['FECHA BAJA'].notnull(), 1, 2)
fecha_baja_category_values = {1: 'Has Value', 2: 'No Value'}
categorization['FECHA BAJA CATEGORY'] = fecha_baja_category_values

# Category for "Nº DE DIAS"
conditions = [
    (df_base['Nº DE DIAS'] <= 30),
    (df_base['Nº DE DIAS'] <= 90),
    (df_base['Nº DE DIAS'] <= 183),
    (df_base['Nº DE DIAS'] <= 365),
    (df_base['Nº DE DIAS'] <= 730),
    (df_base['Nº DE DIAS'] <= 1095),
    (df_base['Nº DE DIAS'] <= 1460),
    (df_base['Nº DE DIAS'] <= 1825),
    (df_base['Nº DE DIAS'] > 1825)
]
choices = [1, 2, 3, 4, 5, 6, 7, 8, 9]
df_base['Nº DE DIAS CATEGORY'] = np.select(conditions, choices)
n_dias_category_values = {
    1: '0-30 days',
    2: '31-90 days',
    3: '91-183 days',
    4: '184-365 days',
    5: '366-730 days',
    6: '731-1095 days',
    7: '1096-1460 days',
    8: '1461-1825 days',
    9: '1826+ days'
}
categorization['Nº DE DIAS CATEGORY'] = n_dias_category_values

# Add the new columns to the categorized DataFrame
df_base_c['FECHA BAJA CATEGORY'] = df_base['FECHA BAJA CATEGORY']
df_base_c['Nº DE DIAS CATEGORY'] = df_base['Nº DE DIAS CATEGORY']

# Print the updated DataFrame and categorized DataFrame
print(df_base_c)
```

```python
# Unir los dataframes utilizando la columna ID como clave
df_machine = df_actual_c[['ID', 'HORAS DE CONTRATO', 'Categoría', 'Número de hijos', 'Sexo', 'Supervisora/Coordinador', 'Sección']].copy()

df_machine = df_machine.merge(df_absentismo_c[['ID', 'IT/AT', 'Horas contratadas', 'TIPO DE BAJA']], on='ID', how='left')
df_machine = df_machine.merge(df_historical_performance_c[['ID', 'Supervisor/a', 'Evaluación']], on='ID', how='left')
df_machine = df_machine.merge(df_performance_c[['ID', 'Evaluación', 'Supervisor/a']], on='ID', how='left')
df_machine = df_machine.merge(df_base_c[['ID', 'CATEGORIA', 'SECCIÓN', 'TIENDA/SECCION INICIAL', 'SECCIÓN2', 'TIENDA/SECCION FINAL2', 'QUIEN CAUSO BAJA?', 'FECHA BAJA CATEGORY', 'Nº DE DIAS CATEGORY']], on='ID', how='left')

# Si la evaluación está vacía en df_performance_c, buscar en df_historical_performance_c
df_machine['Evaluación'] = df_machine['Evaluación_x'].fillna(df_machine['Evaluación_y'])

# Si el supervisor/a está vacío en df_performance_c, buscar en df_historical_performance_c
df_machine['Supervisor/a'] = df_machine['Supervisor/a_x'].fillna(df_machine['Supervisor/a_y'])

# Eliminar las columnas innecesarias
df_machine = df_machine.drop(['Evaluación_x', 'Evaluación_y', 'Supervisor/a_x', 'Supervisor/a_y'], axis=1)

# Reordenar las columnas
df_machine = df_machine[['ID', 'HORAS DE CONTRATO', 'Categoría', 'Número de hijos', 'Sexo', 'Supervisora/Coordinador', 'Sección',
                         'IT/AT', 'Horas contratadas', 'TIPO DE BAJA', 'Evaluación', 'Supervisor/a',
                         'CATEGORIA', 'SECCIÓN', 'TIENDA/SECCION INICIAL', 'SECCIÓN2', 'TIENDA/SECCION FINAL2', 'QUIEN CAUSO BAJA?',
                         'FECHA BAJA CATEGORY', 'Nº DE DIAS CATEGORY']]

# Imprimir el DataFrame df_machine
print(df_machine)
```

```python
# Reemplazar NaN por 9 en las columnas 'IT/AT', 'TIPO DE BAJA' y 'QUIEN CAUSO BAJA?'
df_machine['IT/AT'] = df_machine['IT/AT'].fillna(9)
df_machine['TIPO DE BAJA'] = df_machine['TIPO DE BAJA'].fillna(9)
df_machine['QUIEN CAUSO BAJA?'] = df_machine['QUIEN CAUSO BAJA?'].fillna(9)
df_machine = df_machine.dropna(subset=['Evaluación', 'Supervisor/a', 'Horas contratadas', 'CATEGORIA'])
df_machine = df_machine.drop_duplicates(subset='ID', keep='first')
df_machine = df_machine.set_index('ID')
```
