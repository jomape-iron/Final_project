```python
numero_trabajadores = df_base['FECHA BAJA'].isnull().sum()

print("Número de trabajadores actuales:", numero_trabajadores)
```

```python
# Contar el número de trabajadores sin fecha de baja por categoría
trabajadores_por_categoria = df_base[df_base['FECHA BAJA'].isnull()].groupby('CATEGORIA').size()

print("Número de trabajadores por categoría:")
print(trabajadores_por_categoria)

# Contar el número de trabajadores sin fecha de baja por sección
trabajadores_por_seccion = df_base[df_base['FECHA BAJA'].isnull()].groupby('SECCIÓN').size()

print("Número de trabajadores por sección:")
print(trabajadores_por_seccion)
```

```python
# Obtener el año de la fecha de alta
df_base['Año de alta'] = pd.to_datetime(df_base['FECHA ALTA']).dt.year

# Contar el número de trabajadores dados de alta por año
trabajadores_por_año = df_base.groupby('Año de alta').size()

print("Número de trabajadores dados de alta por año:")
print(trabajadores_por_año)
```

```python
# Convertir la columna "FECHA BAJA" al formato de fecha
df_base['FECHA BAJA'] = pd.to_datetime(df_base['FECHA BAJA'])

# Obtener el año de la fecha de baja
df_base['Año de baja'] = df_base['FECHA BAJA'].dt.year

# Contar el número de trabajadores dados de baja por año
trabajadores_por_año_baja = df_base.groupby('Año de baja').size()

print("Número de trabajadores dados de baja por año:")
print(trabajadores_por_año_baja)
```

```python
# Convertir las columnas 'FECHA ALTA' y 'FECHA BAJA' al formato de fecha
df_base['FECHA ALTA'] = pd.to_datetime(df_base['FECHA ALTA'])
df_base['FECHA BAJA'] = pd.to_datetime(df_base['FECHA BAJA'])

# Filtrar las personas que tienen una fecha de alta pero no una fecha de baja entre el 2019/01/01 y el 2019/12/31
personas_activas_2019 = df_base[(df_base['FECHA ALTA'].notnull()) & (df_base['FECHA BAJA'].isnull()) & (df_base['FECHA ALTA'].dt.year == 2019)]

# Calcular el valor de X para el año 2019
x_2019 = len(personas_activas_2019) / 2

# Filtrar las personas que tienen una fecha de baja correspondiente al año 2019
personas_dadas_de_baja_2019 = df_base[(df_base['FECHA BAJA'].notnull()) & (df_base['FECHA BAJA'].dt.year == 2019)]

# Calcular el valor de Y para el año 2019
y_2019 = len(personas_dadas_de_baja_2019)

# Calcular la rotación de personal para el año 2019
rotacion_personal_2019 = (y_2019 / x_2019) * 1000

print("Rotación de personal en el año 2019:")
print(rotacion_personal_2019)
```
